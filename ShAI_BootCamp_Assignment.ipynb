## pie chart represents the destribution of employees in different departements

âœˆ will add a new column that indicates which departement that employee belogs to

```python
def extract_department(job_title):
    if '(' in job_title and ')' in job_title:
        department = job_title.split('(')[-1].strip(')')
    else:
        department = 'No department specified'
    return department

df['department'] = df['JobTitle'].apply(extract_department)
```

âœˆ convert any uppercase written departement into lowercase

```python
df['department'] = df['department'].map(str.lower)
```

âœˆ replace cvl&crmnl --> civil&criminal

```python
df['department'] = df['department'].str.replace('cvl&crmnl','civil & criminal')
df['department'] = df['department'].str.replace('civil/criminal','civil & criminal')
```

âœˆ replace fire dept--> fire department

```python
df['department'] = df['department'].str.replace('fire dept','fire department')
```

âœˆ Exclude 'No department specified' category

```python
df_filtered = df[df['department'] != 'no department specified']
```

ðŸ¥§ it's time for pie chart

```python
# Pie chart of departments
department_counts = df_filtered['department'].value_counts()
colors = ['red', 'blue', 'green', 'yellow', 'orange','brown','black']  # Specify colors for each department
plt.pie(department_counts, labels=department_counts.index, colors=colors)
plt.title('Proportion of Employees in Different Departments')
plt.axis('equal')
plt.show()
```

# 5. **Grouped Analysis**

## Group by department and calculate average salary

```python
average_salary_by_department = df.groupby('department')['TotalPay'].mean()
```

## ðŸ“Š chart of average salaries by departement

```python
average_salary_by_department.plot(kind='bar')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.title('Average Salaries by Department')
plt.xticks(rotation=45)
plt.show()
```

```python
average_salary_by_department.describe()
```

## group by job title and calculate the average salries

```python
# Define the columns for grouping
group_jobTtitle = ['JobTitle']

# Calculate summary statistics for each group
grouped_jobTtitle = df.groupby(group_jobTtitle).agg({
    'BasePay': 'mean',
    'OvertimePay': 'mean',
    'OtherPay': 'mean',
    'Benefits': 'mean',
    'TotalPay': 'mean',
    'TotalPayBenefits': 'mean'
})

# Print the grouped statistics
grouped_jobTtitle
```

# 6. **Simple Correlation Analysis**

```python
import seaborn as sns
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

## correlation matrix (heatmap style)

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True,)
```

## correlation matrix (zoomed heatmap style)

```python
k = 6 #number of variables for heatmap
cols = corrmat.nlargest(k, 'TotalPay')['TotalPay'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

## Scatter plots between **TotalPay** and correlated variables

```python
sns.set()
cols = ['TotalPay', 'BasePay', 'TotalPayBenefits', 'Benefits', 'OvertimePay', 'OtherPay']
sns.pairplot(df[cols], size = 2.5)
plt.show();
```
